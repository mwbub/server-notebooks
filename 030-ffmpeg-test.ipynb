```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
```

```python
x = np.linspace(-np.pi, np.pi, 100)
```

```python
def f(x, t):
    return np.sin(x - 2*np.pi*t)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ln = plt.plot([], [], animated=True)[0]

def animate(t):
    ln.set_data(x, f(x, t))
    ax.relim()
    ax.autoscale_view()
    
ani = animation.FuncAnimation(fig, animate, frames=np.linspace(0,1,1000), interval=1000/120) 
```

```python
HTML(ani.to_html5_video())
```

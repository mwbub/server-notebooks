```python
import os
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
```

```python
filename = os.path.expandvars('$SCRATCH/data/qdf_samples/200000000_samples.fits')
data = Table.read(filename, format='fits')
```

```python
R, vR, vT, z, vz, phi = data['R', 'vR', 'vT', 'z', 'vz', 'phi'].as_array().view((np.float64, 6)).T
del data
```

```python
mask = (R > 3/8.) & (R < 13/8.)
plt.hist2d(R[mask], vT[mask], bins=500)
plt.show()
```

```python
X = R * np.cos(phi)
Y = R * np.sin(phi)
```

```python
plt.hist2d(X, Y, bins=1000)
plt.show()
```
